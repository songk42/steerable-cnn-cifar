```python
def plot_confusion_matrix(matrix, title='Confusion matrix'):
    fig, ax = plt.subplots()
    im  = ax.imshow(matrix)
    ax.set_xticks(np.arange(10), labels=('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck'))
    ax.set_yticks(np.arange(10), labels=('plane-pred', 'car-pred', 'bird-pred', 'cat-pred',
           'deer-pred', 'dog-pred', 'frog-pred', 'horse-pred', 'ship-pred', 'truck-pred'))
    ax.set_title(title)
    for i in range(10):
        for j in range(10):
            percent = matrix[i, j] / matrix[i].sum() * 100
            text = ax.text(
                j, i,
                f"{matrix[i, j]}",
                # f"{percent:0.2f}%\n{matrix[i, j]}",
                ha='center', va='center', color='w')
```
